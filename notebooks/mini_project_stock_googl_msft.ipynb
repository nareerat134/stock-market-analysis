# **GOOGLE AND MICROSOFT CORPORATION STOCK ANALYSIS**

# **Data Fetching - Short Term Analysis (20 Days)**

```python
ticker1_symbol = "GOOGL" #defines the ticker symbol of the company we want to analyze
ticker2_symbol = "MSFT"
```

```python
stock1_data = yf.Ticker(ticker1_symbol) #gets the stock data for Google
stock2_data = yf.Ticker(ticker2_symbol) #gets the stock data for Microsoft Corporation
```

```python
hist1_data = stock1_data.history(period="20d")# defines what period we want to get the data for; shorterm 20 days
hist2_data = stock2_data.history(period="20d")
```

```python
display(hist1_data.head()) #displays the stock data in a table with the headings
display(hist2_data.head()) #displays the stock data in a table with the headings
```

```python
hist1_data['Daily Return'] = hist1_data['Close'].pct_change() #Adds a daily return column to the table
hist2_data['Daily Return'] = hist2_data['Close'].pct_change()
```

# **Data Cleaning**

```python
hist1_data.fillna('-', inplace=True)# Replace NaN values with a hyphen ('-') using .fillna()
display(hist1_data)# Display the new table after replacing NaN values
hist2_data.fillna('-', inplace=True)# Replace NaN values with a hyphen ('-') using .fillna()
display(hist2_data)# Display the new table after replacing NaN values
```

# **Data Visualization - 20 days**

```python
import matplotlib.pyplot as plt #imports a sub library within matplotlib to help with visualizations
```

```python
%matplotlib inline
```

```python
#Creates a plot for the stock data over 20 days
plt.figure(figsize=(10,5)) #determines the size of the graph
plt.plot(hist1_data['Close'],color = 'red', label='GOOGL Stock Over 20 days') #sets the information to be plotted, the legend label and the clolor of the line
plt.plot(hist2_data['Close'],color = 'green', label='MSFT Stock Over 20 days') #sets the information to be plotted, the legend label and the clolor of the line
plt.title('Clsoing Price Over 20 days') #Adds the title to the graph
plt.xlabel('Closing Date') #Labels the x axis
plt.ylabel('Closing price USD ($)') #Labels the y axis
plt.legend() #Adds a legend to the plot
plt.xticks(rotation=45)#Rotates the x axis so we can see the infromation more clearly
plt.tight_layout() #Adjust layout to prevent overlapping of any labels
plt.show() #Displays the plot
```

```python
#Plots the volume traded in the last 20 days

volume1_traded = hist1_data['Volume'] #gets what infromation needs to be plotted
volume2_traded = hist2_data['Volume']
plt.figure(figsize=(10,5))# determines the size of the plot
plt.plot(volume1_traded, color='red',label='GOOGL Volume traded over 20 days')#plots the volume trades and sets the color labels the legend
plt.plot(volume2_traded, color='green',label='MSFT Volume traded over 20 days')
plt.xlabel('Date')#Labels the x axis
plt.ylabel('Volume Traded')# Labels the y axis
plt.title('Volume Traded Over 20 Days')# Adds the title of the plot
plt.legend()# Adds a legend to the plot
plt.xticks(rotation=45)# Rotates the x axis so we can see the infromation more clearly
plt.tight_layout() # Adjust layout to prevent overlapping of any labels
plt.show() #Displays the plot
```

```python
# Plots the 30 and 300 days moving average for the last 20 days

hist1_data['30_Move_Avg'] = hist1_data['Close'].rolling(window=30).mean() # Calculates the 30 day moving average for the closing prices over the last 20 days
hist2_data['30_Move_Avg'] = hist1_data['Close'].rolling(window=30).mean()
hist1_data['300_Move_Avg'] = hist1_data['Close'].rolling(window=300).mean() # Calculates the 300 day moving average for the closing prices over the last 20 days
hist2_data['300_Move_Avg'] = hist2_data['Close'].rolling(window=300).mean()
plt.figure(figsize=(10, 5)) #determines the size of the plot
plt.plot(hist1_data['Close'], label='GOOGL Close Price',color='red')#Plots the  closing prices from the last 20 days,sets the label for the legend and sets the color of the line
plt.plot(hist1_data['30_Move_Avg'], label='GOOGL 30 day Moving Average',color='orange')#Plots the  30 day moving average,sets the label for the legend and sets the color of the line
plt.plot(hist1_data['300_Move_Avg'], label='GOOGL 300 Moving Average',color='yellow')#Plots the  300 day moving average,sets the label for the legend and sets the color of the line
plt.plot(hist2_data['Close'], label='MSFT Close Price',color='green')
plt.plot(hist2_data['30_Move_Avg'], label='MSFT 30 day Moving Average',color='violet')
plt.plot(hist2_data['300_Move_Avg'], label='MSFT 300 Moving Average',color='blue')
plt.xlabel('Date') # Labels the x-axis
plt.ylabel('Price') # Labels the y-axis
plt.title('Moving Averages of GOOL and MSFT Stock over the last 20 days ') # Adds the plot title
plt.legend()# Adds a legend
plt.xticks(rotation=45)# Rotates the x axis so we can see the infromation more clearly
plt.tight_layout() # Adjust layout to prevent overlapping of any labels
plt.show()# Displays the  plot
```

```python
import matplotlib.pyplot as plt

# Data
companies = ['GOOGL', 'MSFT']
min_price = [min1_price, min2_price]  # Minimum price for each company
max_price = [max1_price, max2_price]  # Maximum price for each company
percentage_change = [percentage1_change, percentage2_change]  # Percentage change for each company
average_price = [average1_price, average2_price]  # Average price for each company
median_price = [median1_price, median2_price]  # Median price for each company

# Plotting
x = range(len(companies))  # X-axis positions
bar_width = 0.15  # Width of each bar

plt.bar(x, min_price, width=bar_width, label='Min Price')
plt.bar([i + bar_width for i in x], max_price, width=bar_width, label='Max Price')
plt.bar([i + 2*bar_width for i in x], percentage_change, width=bar_width, label='Percentage Change')
plt.bar([i + 3*bar_width for i in x], average_price, width=bar_width, label='Average Price')
plt.bar([i + 4*bar_width for i in x], median_price, width=bar_width, label='Median Price')

# Customizing the plot
plt.xlabel('Company')
plt.ylabel('Price')
plt.title('Comparison of Stock Metrics between GOOGL and MSFT')
plt.xticks([i + 2*bar_width for i in x], companies)
plt.legend()
plt.tight_layout()

# Show plot
plt.show()
```
